# Seaborn圖樣式
有五種預設seaborn主題：darkgrid，whitegrid，dark，white，和ticks。它們各自適合不同的應用和個人喜好。默認主題是darkgrid

```python
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity="all"
```

```python
import numpy as np
import seaborn as sns

#設定圖形樣式 - whitegrid
# use sns.set
sns.set_style("whitegrid")
# 利用 NUMPY 去建立資料集
# np.random.RandomState 設定數學式
rs = np.random.RandomState(4) #確保每次圖片輸出一樣
x = rs.normal(2, 1, 75)
y = 2 + 1.5 * x + rs.normal(0, 2, 75)

# 畫圖
sns.residplot(x,y,lowess=True)
```

繪製單變數分佈: 
快速檢視海出生單變數分佈的最便捷方式是distplot() 函數。默認情況下,這將繪製直方圖並適合內核密度估計值(KDE)。

作業目標(2)
(1)使用 distplot()使用簡單的規則來正確猜測預設情況下正確的數位,但嘗試更多或更少的 bin 可能會顯示資料中的其他特徵:
(2)有無kde對圖形分布的影響

```python
# bin: 指的是特徵值, 
# kde: on/off
# sns.distplot();
import matplotlib.pyplot as plt
sns.set(rc={'figure.figsize':(22.7,11.7)})
sns.set(color_codes=True)
sns.set_style("whitegrid")
#x = np.random.normal(size=100) 
#default displot
fig, ax =plt.subplots(1,2)
sns.distplot(x, kde=False, rug=True, ax=ax[0]).set_title("default displot")
sns.distplot(x, kde=True, rug=True, ax=ax[1]).set_title("default displot with kde")
fig.show()
```

```python
fig,ax = plt.subplots(1,2)
sns.distplot(x, bins=20, kde=False, rug=True,ax=ax[0]).set_title("distplot with bins=20")
sns.distplot(x, bins=20, kde=True, rug=True, ax=ax[1]).set_title("distplot with bins=20 and kde")
fig.show()
#using more bins can help you know the maximum and minimum more clearly 
```

```python
fig,ax = plt.subplots(1,2)
sns.distplot(x, bins=4, kde=False, rug=True, ax=ax[0]).set_title("distplot with bins=4") 
sns.distplot(x, bins=4, kde=True, rug=True, ax=ax[1]).set_title("distplot with bins=4 and kde")
fig.show()
#less bins will enhance the effect of mean
```
